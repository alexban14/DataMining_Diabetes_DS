# 2. Logistic Regression

- applying the Logistic Regression Model on the dataset

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

### Spliting the dataset into training and testing sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
```

### Training the model using the training data and utilizing different regularization parameters

```python
# list of regularization parameter values
C_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

# parameter grid with L2 regularization
param_grid = {'C': C_values, 'penalty': ['l2'], 'solver': ['liblinear']}

LogisticRegressor = LogisticRegression(max_iter = 1000)

grid_search = GridSearchCV(LogisticRegressor, param_grid, cv = 5, scoring = 'accuracy')
grid_search.fit(X_train, Y_train)

print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score: ", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
best_model.fit(X_train, Y_train)

print("Regression coefficients obtained :")
print("The intercept: ", best_model.intercept_)
print("*The coefficient: ", best_model.coef_)
```

## Model evaluation

### Accuracy score on the training data

```python
X_train_prediction = best_model.predict(X_train)
training_score_accuracy = accuracy_score(Y_train, X_train_prediction)

print('Accuracy score of training data: ', training_score_accuracy)
```

### Accuracy score on the test data

```python
X_test_prediction = best_model.predict(X_test)
training_score_accuracy = accuracy_score(Y_test, X_test_prediction)

print('Accuracy score of test data: ', training_score_accuracy)
```

# 4. Associtation Rules using the Apriori algorithm

## Data Preprocessing

- transforming the continuous features into descrete classes

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Binning continuous features into categories
diabetes_dataset['Pregnancies'] = pd.cut(diabetes_dataset['Pregnancies'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
diabetes_dataset['Glucose'] = pd.cut(diabetes_dataset['Glucose'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
diabetes_dataset['BloodPressure'] = pd.cut(diabetes_dataset['BloodPressure'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
diabetes_dataset['SkinThickness'] = pd.cut(diabetes_dataset['SkinThickness'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
diabetes_dataset['Insulin'] = pd.cut(diabetes_dataset['Insulin'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
diabetes_dataset['BMI'] = pd.cut(diabetes_dataset['BMI'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
diabetes_dataset['DiabetesPedigreeFunction'] = pd.cut(diabetes_dataset['DiabetesPedigreeFunction'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
diabetes_dataset['Age'] = pd.cut(diabetes_dataset['Age'], bins=5, labels=['Very Young', 'Young', 'Middle Age', 'Old', 'Very Old'])


# Convert the Outcome feature to a categorical type
diabetes_dataset['Outcome'] = diabetes_dataset['Outcome'].astype('str')

# Print the modified DataFrame
print(diabetes_dataset.head())
```

```python
# Convert the DataFrame to a one-hot encoded format
one_hot = pd.get_dummies(diabetes_dataset)
print(one_hot.head())
```

```python
# Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(one_hot, min_support=0.1, use_colnames=True)
print(frequent_itemsets)

# Generate association rules from the frequent itemsets
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
print(rules)
```

### Analyzing the Results

- interpreting the association rules to find meaningful patterns.

```python
rules = rules.sort_values('lift', ascending=False)
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

### Support, Confidence & Lift

```python
support = rules['support']
confidence = rules['confidence']
lift = rules['lift']

# Plotting the support, confidence, and lift
plt.figure(figsize=(12, 8))
plt.scatter(support, confidence, alpha=0.5, c=lift, cmap='viridis')
plt.colorbar(label='Lift')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence Colored by Lift')
plt.show()
```

### Lift vs. Confidence

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='confidence', y='lift', size='support', hue='support', data=rules)
plt.title('Lift vs Confidence')
plt.xlabel('Confidence')
plt.ylabel('Lift')
plt.legend()
plt.show()
```

### Association Rules Heatmap

```python
top_rules = rules.nlargest(20, 'lift')

heatmap_data = top_rules.pivot(index='antecedents', columns='consequents', values='support')

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='viridis')
plt.title('Heatmap of Support for Association Rules')
plt.show()
```

## Clustering

## Partition Clustering

- fundamental technique in unsupervised machine learning used to group similar data points into clusters or partitions.
- it aims to discover natural groupings or clusters within a dataset, where data points within the same cluster are more similar to each other than to those in other clusters

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score, adjusted_rand_score
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### K-Means Clustering

```python
# Initialize the K-means clustering algorithm with a specific k
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the data
kmeans.fit(X_scaled)

# Get the cluster assignments for each sample
kmeans_labels = kmeans.labels_

# Calculate the silhouette score
kmeans_silhouette_score = silhouette_score(X_scaled, kmeans_labels)

# Calculate the adjusted Rand Index (ARI) score
kmeans_ari_score = adjusted_rand_score(Y, kmeans_labels)
print(f"K-means Clustering:")
print(f"Silhouette Score: {kmeans_silhouette_score:.4f}")
print(f"Adjusted Rand Index Score: {kmeans_ari_score:.4f}")
```

### K-Meloids Clustering

```python
# Initialize the K-medoids clustering algorithm with a specific k
kmedoids = KMedoids(n_clusters=3, random_state=42)

# Fit the model to the data
kmedoids.fit(X_scaled)

# Get the cluster assignments for each sample
kmedoids_labels = kmedoids.labels_

# Calculate the silhouette score
kmedoids_silhouette_score = silhouette_score(X_scaled, kmedoids_labels)
```

```python
# Calculate the adjusted Rand Index (ARI) score
kmedoids_ari_score = adjusted_rand_score(Y, kmedoids_labels)

print(f"\nK-medoids Clustering:")
print(f"Silhouette Score: {kmedoids_silhouette_score:.4f}")
print(f"Adjusted Rand Index Score: {kmedoids_ari_score:.4f}")
```

### Comparing the 2 algorithms used

```python
# Visualization of K-means and K-medoids Clusters
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1]
    , c=kmeans_labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0]
    , kmeans.cluster_centers_[:, 1]
    , marker='X', s=200, c='red', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Sepal Length (Scaled)')
plt.ylabel('Sepal Width (Scaled)')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0]
    , X_scaled[:, 1], c=kmedoids_labels, cmap='viridis')
plt.scatter(kmedoids.cluster_centers_[:, 0]
    , kmedoids.cluster_centers_[:, 1]
    , marker='X', s=200, c='red', label='Medoids')
plt.title('K-medoids Clustering')
plt.xlabel('Sepal Length (Scaled)')
plt.ylabel('Sepal Width (Scaled)')
plt.legend()

plt.tight_layout()
plt.show()
```
