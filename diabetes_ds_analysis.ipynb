## Standard Deviation Plots

### Every feature std. deviation for Outcome 1

```python
# Split the dataset into features (X) and target variable (Y)
X = diabetes_dataset.drop('Outcome', axis=1)
Y = diabetes_dataset['Outcome']

# Calculate standard deviation between each feature and the outcome
std_dev_1 = []
for column in X.columns:
    std_dev_1.append(np.std(X[column][Y == 1]))  # Calculate standard deviation for outcome 1

# Plot the standard deviation curve
plt.figure(figsize=(10, 6))
plt.plot(X.columns, std_dev_1) # , marker='o', linestyle='-'
plt.xticks(rotation=45)
plt.xlabel('Features')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviation of the dataset (1)')
plt.grid(True)
plt.show()
```

# Logistic Regression

- applying the simple Logistic Regression Model on our dataset

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

### Spliting the dataset into training and testing sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
```

### Training the model using the training data and utilizing different regularization parameters

```python
# list of regularization parameter values
C_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

# parameter grid with L2 regularization
param_grid = {'C': C_values, 'penalty': ['l2'], 'solver': ['liblinear']}

LogisticRegressor = LogisticRegression(max_iter = 1000)

grid_search = GridSearchCV(LogisticRegressor, param_grid, cv = 5, scoring = 'accuracy')
grid_search.fit(X_train, Y_train)

print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score: ", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
best_model.fit(X_train, Y_train)

print("Regression coefficients obtained :")
print("The intercept: ", best_model.intercept_)
print("*The coefficient: ", best_model.coef_)
```

## Model evaluation

### Accuracy score on the training data

```python
X_train_prediction = best_model.predict(X_train)
training_score_accuracy = accuracy_score(Y_train, X_train_prediction)

print('Accuracy score of training data: ', training_score_accuracy)
```

### Evaluating the model using a confusion matrix on the training data

```python
ConfusionMatrix = confusion_matrix(Y_train, X_train_prediction, labels = [1, 0])
TP, FN, FP, TN = ConfusionMatrix.ravel()

print(f"TP: {TP}, FN: {FN}, FP: {FP}, TN: {TN}")

conf_mx_df = pd.DataFrame(
    [
        [TN, FP],
        [FN, TP],
    ],
    index = ['Actual 0', 'Actual 1'],
    columns = ['Predicted 0', 'Predicted 1']
)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_mx_df, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={"size": 16})
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

### Accuracy score on the test data

```python
X_test_prediction = best_model.predict(X_test)
training_score_accuracy = accuracy_score(Y_test, X_test_prediction)

print('Accuracy score of test data: ', training_score_accuracy)
```

### Evaluating the model using a confusion matrix on the test data

- this helps to obtain an accuracy rate

```python
ConfusionMatrix = confusion_matrix(Y_test, X_test_prediction, labels = [1, 0])
TP, FN, FP, TN = ConfusionMatrix.ravel()

print(f"TP: {TP}, FN: {FN}, FP: {FP}, TN: {TN}")

conf_mx_df = pd.DataFrame(
    [
        [TN, FP],
        [FN, TP],
    ],
    index = ['Actual 0', 'Actual 1'],
    columns = ['Predicted 0', 'Predicted 1']
)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_mx_df, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={"size": 16})
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

### Classification Report

```python
classification_report_result = classification_report(Y_test, X_test_prediction, target_names=['No Diabetes', 'Diabetes'])

print('Classification Report:')
print(classification_report_result)
```
