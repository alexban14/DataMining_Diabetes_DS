## Standard Deviation Plots

### Every feature std. deviation for Outcome 1

```python
# Split the dataset into features (X) and target variable (Y)
X = diabetes_dataset.drop('Outcome', axis=1)
Y = diabetes_dataset['Outcome']

# Calculate standard deviation between each feature and the outcome
std_dev_1 = []
for column in X.columns:
    std_dev_1.append(np.std(X[column][Y == 1]))  # Calculate standard deviation for outcome 1

# Plot the standard deviation curve
plt.figure(figsize=(10, 6))
plt.plot(X.columns, std_dev_1) # , marker='o', linestyle='-'
plt.xticks(rotation=45)
plt.xlabel('Features')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviation of the dataset (1)')
plt.grid(True)
plt.show()
```

## Data Distribution Plot

```python
plt.figure(figsize=(15, 10))
for i, column in enumerate(diabetes_dataset.columns):
    plt.subplot(3, 3, i + 1)
    plt.hist(diabetes_dataset[column], bins='auto')
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Count')
    plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.set(style="whitegrid")
for i, column in enumerate(X.columns):
    plt.subplot(3, 3, i + 1)
    sns.histplot(X[column], kde=True)
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Count')
    
plt.tight_layout()
plt.show()
```
